## Functions

It is a way to group some code under one name

Running the function's code = calling the function

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_columns(df, col_name, separate_plots=False):
    '''
    This is a function that makes a plot of a particular column in a dataframe.
    This is a cool function. This is called a docstring.
    '''
    plt.plot(range(len(df[col_name])), df[col_name], label=col_name)
    print(col_name)
    plt.legend()
    if separate_plots:
        plt.show()
```

```python
df = pd.read_csv("motorola.csv")
```

```python
df.head()
```

```python
plot_columns(df, "Revenue (in millions)")
```

```python
for column in df.drop(["Year", "Company"], axis=1).columns:
    plot_columns(df, column, separate_plots=True)
```
