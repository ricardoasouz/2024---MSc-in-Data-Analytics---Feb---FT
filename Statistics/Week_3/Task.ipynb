###Demonstrate the Central Limit Theorem (CLT) using the built-in "tips" dataset from seaborn, we'll simulate random samples of the total bill amount and observe how the distribution of sample means approaches a normal distribution as the sample size increases.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

Step1:Import Libraries

Step2: Load the tips dataset. tips = sns.load_dataset("tips")

Step3: Extract the total bill amount column

Step4: Define parametrs for sample size and number of samples

Step5: Perform sampling with diffrent sample size and plot the result till you get the bell shape distribution


```python
tips = sns.load_dataset("tips")
```

```python
total_bill = tips['total_bill']
tip = tips['tip']
```

```python
# Plot the distribution of the population data
plt.figure(figsize=(10, 6))
plt.hist(total_bill, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title('Tips')
plt.xlabel('Total_Bill')
plt.ylabel('Number of cases')

plt.show()
```

```python
#import numpy as np
#import matplotlib.pyplot as plt

# Define the range of sample sizes and number of samples
sample_sizes = [10, 20, 30, 40, 50]  # Different sample sizes
num_samples = 1000  # Number of samples

# Create a figure to plot the distributions
plt.figure(figsize=(12, 8))

# Iterate over different sample sizes
for size in sample_sizes:
    # Initialize an array to store the means of each sample
    means = []
    
    # Generate samples and calculate the mean for each sample
    for _ in range(num_samples):
        sample = np.random.choice(tips['total_bill'], size=size, replace=False)
        means.append(np.mean(sample))
    
    # Plot the histogram of sample means
    plt.hist(means, bins=30, alpha=0.6, label=f'Sample Size: {size}')

# Add labels and title to the plot
plt.xlabel('Mean of Samples')
plt.ylabel('Frequency')
plt.title('Sampling Distribution with Different Sample Sizes')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
