```python
# Ensure that all dataframes have the same length
if not (len(df_production) == len(df_commodity_price) == len(df_consumer_price) == len(df_import_price)):
    print("DataFrames do not have the same number of rows.")
else:
    # Compare index by index
    for index in range(len(df_production)):
        time_periods = [df_production.loc[index, 'TIME_PERIOD'], 
                        df_commodity_price.loc[index, 'TIME_PERIOD'], 
                        df_consumer_price.loc[index, 'TIME_PERIOD'], 
                        df_import_price.loc[index, 'TIME_PERIOD']]
        
        geos = [df_production.loc[index, 'geo'], 
                df_commodity_price.loc[index, 'geo'], 
                df_consumer_price.loc[index, 'geo'], 
                df_import_price.loc[index, 'geo']]
        
        if not (time_periods[0] == time_periods[1] == time_periods[2] == time_periods[3]):
            print(f"Mismatch in 'TIME_PERIOD' at index {index}: {time_periods}")
        
        if not (geos[0] == geos[1] == geos[2] == geos[3]):
            print(f"Mismatch in 'geo' at index {index}: {geos}")

    print("Comparison complete.")
```

```python
#CREATING A NEW DATA FRAME, FIRST CLEANING PRODUCTION DATASET THEN MERGING OBS_VALUE FROM THE PRICE'S DATASET'S

# List of columns to remove
columns_to_remove = ['DATAFLOW', 'LAST UPDATE', 'freq', 'meat', 'meatitem', 'unit', 'OBS_FLAG']

# Remove the specified columns
df = df_production.drop(columns=columns_to_remove)

#Renaming OBS_VALUE to prod
df.rename(columns={'OBS_VALUE': 'prod'}, inplace=True)
```

```python
df
```

```python
#Merging the OBS_VALUE of datasets on new data frame DF

# Define a list of tuples containing the dataset and the new column name
datasets = [
    (df_commodity_price, 'commodity'),
    (df_consumer_price, 'consumer'),
    (df_import_price, 'import')
]

# Merge each dataset with the main DataFrame 'df'
for dataset, new_column in datasets:
    df = df.merge(
        dataset[['TIME_PERIOD', 'geo', 'OBS_VALUE']].rename(columns={'OBS_VALUE': new_column}),
        on=['TIME_PERIOD', 'geo'],
        how='left'
    )

# Display the first few rows of the merged DataFrame
df.head()
```

```python
#!pip install pycountry
import pycountry

# Custom dictionary for country codes to country names
custom_country_mapping = {
    'EL': 'Greece',
    # Add other custom mappings as needed
}

# Function to convert country code to country name using both pycountry and custom mapping
def combined_get_country_name(code):
    if code in custom_country_mapping:
        return custom_country_mapping[code]
    try:
        return pycountry.countries.get(alpha_2=code).name
    except AttributeError:
        return code

# Apply the function to the 'geo' column
df['geo'] = df['geo'].apply(combined_get_country_name)

# Display the first few rows of the updated dataset
df.head()
```

```python
df.info()
```

```python
# Pivot the DataFrame by 'TIME_PERIOD' and rename the columns as requested
pivot_df = df.pivot(index='TIME_PERIOD', columns='geo', values=['prod', 'commodity', 'consumer', 'import'])

# Flatten the MultiIndex columns and rename them
pivot_df.columns = [f'{geo}_{metric}' for metric, geo in pivot_df.columns]

# Sort the columns alphabetically
pivot_df = pivot_df.reindex(sorted(pivot_df.columns), axis=1)
```

```python
import dash
from dash import dcc, html, Input, Output, State
import dash_table
import pandas as pd
import plotly.graph_objects as go
from scipy.stats import ttest_ind, f_oneway, chi2_contingency, ranksums, mannwhitneyu, pearsonr
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, LogisticRegression
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, silhouette_score, accuracy_score
import numpy as np
import statsmodels.api as sm
from statsmodels.miscmodels.ordinal_model import OrderedModel

# Assuming pivot_df is already loaded into memory

# Extract feature columns and target columns
features = [col for col in pivot_df.columns if any(metric in col for metric in ['commodity', 'import', 'consumer', 'prod'])]

# Preprocess the dataset
X = pivot_df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

def perform_ml_tasks(selected_countries):
    if not selected_countries:
        return [], {}

    results = []

    # Filter features for the selected countries
    features = [col for col in pivot_df.columns if any(metric in col for metric in ['commodity', 'import', 'consumer']) and col.split('_')[0] in selected_countries]
    targets = [col for col in pivot_df.columns if any(metric in col for metric in ['prod', 'commodity', 'import', 'consumer']) and col.split('_')[0] in selected_countries]

    if not features or not targets:
        return [], {}

    X = pivot_df[features]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    predictions = {}
    
    for target in targets:
        y = pivot_df[target]

        # Train and predict with Linear Regression
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
        lin_reg = LinearRegression()
        lin_reg.fit(X_train, y_train)
        y_pred_lin_reg = lin_reg.predict(X_test)
        predictions[target] = (y_test, y_pred_lin_reg)

        mse_lin_reg = mean_squared_error(y_test, y_pred_lin_reg)
        lin_reg_scores = cross_val_score(lin_reg, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
        lin_reg_mean_mse = -lin_reg_scores.mean()
        results.append({
            'Model': f'Linear Regression for {target}',
            'Mean Squared Error': mse_lin_reg,
            'Cross-Validation MSE': lin_reg_mean_mse
        })

        # Supervised Learning: Ridge Regression with GridSearchCV
        ridge = Ridge()
        param_grid_ridge = {'alpha': [0.1, 1, 10, 100]}
        grid_search_ridge = GridSearchCV(ridge, param_grid_ridge, cv=5, scoring='neg_mean_squared_error')
        grid_search_ridge.fit(X_train, y_train)
        best_ridge = grid_search_ridge.best_estimator_
        y_pred_ridge = best_ridge.predict(X_test)
        mse_ridge = mean_squared_error(y_test, y_pred_ridge)
        ridge_scores = cross_val_score(best_ridge, X_scaled, y, cv=5, scoring='neg_mean_squared_error')
        ridge_mean_mse = -ridge_scores.mean()
        results.append({
            'Model': f'Ridge Regression for {target}',
            'Mean Squared Error': mse_ridge,
            'Cross-Validation MSE': ridge_mean_mse,
            'Best Parameters': str(grid_search_ridge.best_params_)  # Convert dictionary to string
        })

    # Unsupervised Learning: KMeans Clustering
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans.fit(X_scaled)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    results.append({
        'Model': 'KMeans Clustering',
        'Silhouette Score': silhouette_avg
    })

    return results, predictions

# Create a Dash app
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("European Beef Analyses (slaughterings and prices)"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[
            {'label': 'All', 'value': 'All'},
            {'label': 'None', 'value': 'None'}
        ] + [{'label': country, 'value': country} for country in pivot_df.columns.str.split('_').str[0].unique()],
        value=['All'],
        multi=True,
        placeholder="Select countries"
    ),
    dcc.Dropdown(
        id='metric-dropdown',
        options=[
            {'label': 'All', 'value': 'All'},
            {'label': 'None', 'value': 'None'}
        ] + [{'label': metric, 'value': metric} for metric in ['prod', 'commodity', 'consumer', 'import']],
        value=['All'],
        multi=True,
        placeholder="Select metrics"
    ),
    html.Button('Perform Tests and ML Tasks', id='perform-tasks-button', n_clicks=0),
    dcc.Graph(id='combined-graph'),
    html.H2("Summary Metrics"),
    html.Div(id='summary-table', style={'overflowX': 'scroll'}),
    html.H2("Inferential Statistics"),
    html.Div(id='test-results', style={'display': 'flex', 'justify-content': 'space-between', 'flex-wrap': 'wrap'}),
    html.H2("Machine Learning Tasks"),
    html.Div(id='ml-results-table', style={'overflowX': 'scroll'})
])

@app.callback(
    [Output('country-dropdown', 'value'),
     Output('metric-dropdown', 'value')],
    [Input('country-dropdown', 'value'),
     Input('metric-dropdown', 'value')]
)
def update_dropdowns(selected_countries, selected_metrics):
    if 'All' in selected_countries:
        all_countries = list(pivot_df.columns.str.split('_').str[0].unique())
        selected_countries = all_countries
    elif 'None' in selected_countries:
        selected_countries = []

    if 'All' in selected_metrics:
        all_metrics = ['prod', 'commodity', 'consumer', 'import']
        selected_metrics = all_metrics
    elif 'None' in selected_metrics:
        selected_metrics = []

    return selected_countries, selected_metrics

@app.callback(
    [Output('combined-graph', 'figure'),
     Output('summary-table', 'children'),
     Output('test-results', 'children'),
     Output('ml-results-table', 'children')],
    [Input('country-dropdown', 'value'),
     Input('metric-dropdown', 'value'),
     Input('perform-tasks-button', 'n_clicks')]
)
def update_combined_graph(selected_countries, selected_metrics, n_clicks):
    if n_clicks > 0 and len(selected_countries) > 4:
        selected_countries = selected_countries[:4]

    fig = go.Figure()
    summary_data = {}

    if 'All' in selected_countries:
        selected_countries = list(pivot_df.columns.str.split('_').str[0].unique())
    elif 'None' in selected_countries:
        selected_countries = []

    if 'All' in selected_metrics:
        selected_metrics = ['prod', 'commodity', 'consumer', 'import']
    elif 'None' in selected_metrics:
        selected_metrics = []

    for country in selected_countries:
        for metric in selected_metrics:
            col = f"{country}_{metric}"
            if col in pivot_df.columns:
                fig.add_trace(go.Scatter(
                    x=pivot_df.index,
                    y=pivot_df[col],
                    mode='lines',
                    name=f"{country} {metric}",
                    hovertemplate=f'<b>{country} {metric}</b><br>Time: %{{x}}<br>Value: %{{y}}<extra></extra>'
                ))
                summary_data[col] = pivot_df[col].describe()

    test_results = []
    ml_results = []
    predictions = {}

    if n_clicks > 0:
        ml_results, predictions = perform_ml_tasks(selected_countries)
        
        for target, (y_test, y_pred) in predictions.items():
            country, metric = target.split('_')
            fig.add_trace(go.Scatter(
                x=pivot_df.index[-len(y_test):],
                y=y_test,
                mode='lines',
                name=f"Actual {country} {metric}",
                hovertemplate=f'<b>Actual {country} {metric}</b><br>Time: %{{x}}<br>Value: %{{y}}<extra></extra>'
            ))
            fig.add_trace(go.Scatter(
                x=pivot_df.index[-len(y_pred):],
                y=y_pred,
                mode='lines',
                name=f"Predicted {country} {metric}",
                line=dict(dash='dash'),
                hovertemplate=f'<b>Predicted {country} {metric}</b><br>Time: %{{x}}<br>Value: %{{y}}<extra></extra>'
            ))

    fig.update_layout(
        title='Combined Actual vs Predicted and Statistical Analysis',
        xaxis_title='Time Period',
        yaxis_title='Values',
        legend_title='Metrics'
    )

    if summary_data:
        summary_df = pd.DataFrame(summary_data).transpose().reset_index()
        summary_table = dash_table.DataTable(
            columns=[{"name": i, "id": i} for i in summary_df.columns],
            data=summary_df.to_dict('records'),
            style_table={'overflowX': 'auto'},
            style_cell={'textAlign': 'left'}
        )
    else:
        summary_table = html.Div("No data available for the selected countries.")

    if ml_results:
        ml_results_df = pd.DataFrame(ml_results)
        ml_table = dash_table.DataTable(
            columns=[{"name": i, "id": i} for i in ml_results_df.columns],
            data=ml_results_df.to_dict('records'),
            style_table={'overflowX': 'auto'},
            style_cell={'textAlign': 'left'}
        )
    else:
        ml_table = html.Div("Perform ML tasks to see results.")

    return fig, summary_table, test_results, ml_table

if __name__ == '__main__':
    app.run_server(debug=True)
```
