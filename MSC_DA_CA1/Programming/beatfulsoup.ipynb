```python
from urllib.request import urlopen
from html.parser import HTMLParser

# A simple HTML parser to find links
class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    print("Found a link:", attr[1])

# URL you want to scrape
url = 'https://data.gov.ie/dataset/?q=tourism&sort=score+desc%2C+metadata_created+desc'

# Open the URL and read the HTML content
response = urlopen(url)
html = response.read().decode('utf-8')

# Initialize our parser and feed it the HTML
parser = MyHTMLParser()
parser.feed(html)
```
